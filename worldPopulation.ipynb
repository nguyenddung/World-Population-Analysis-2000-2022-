## Task 1

```python
most_population = df.sort_values(by="2022 Population",ascending = False)
new_df = df[["Rank","CCA3","Country/Territory","Capital","Continent","2022 Population","Area (km)","Density (per km)","Growth Rate","World Population Percentage"]]
min_population = df.sort_values(by="2022 Population")
# most_population.head(10)
# min_population.head(10)

print("10 quoc gia dong dan nhat la : \n ",most_population.head(10))
print("\n##################\n")
print("10 quoc gia it dan nhat la : \n ",min_population.head(10))
```

## Task 2

```python
### vai lon cach them 1 cot moi
df["Growth_2000_2022(%)"] = ((df["2022 Population"] - df["2000 Population"]) 
                              / df["2000 Population"]) * 100
new_df = df.sort_values(by="Growth_2000_2022(%)",ascending= False)
new_df = new_df[["Rank","Country/Territory","Growth_2000_2022(%)"]]
new_df.head(10)
```

## Task 3

```python
new_df = df.sort_values(by="Density (per km)",ascending= False)

new_df.head(10)
```

## Task4

```python
new_df_sum_population_2022 = df.groupby("Country/Territory")["2022 Population"].sum()
# new_df_sum_population_2022
new_df_sum_population_2022_sorted = new_df_sum_population_2022.sort_values(ascending= False)
new_df_sum_population_2022_sorted
```

## Task 5

```python
total_dan_so_2022 = 7990000000
df["ti_le_dan_so"] = ((df["2022 Population"] / total_dan_so_2022) * 100).round(10)
df_sort = df.sort_values(by="ti_le_dan_so",ascending= False)
df_sort
```

## Task 10

```python
# Tnh h s tng quan gia Area v Population 2022
corr_area_population = df["Area (km)"].corr(df["2022 Population"])

# Tnh h s tng quan gia Population 2022 v Density
corr_population_density = df["2022 Population"].corr(df["Density (per km)"])

print("H s tng quan gia Din tch v Dn s 2022:", round(corr_area_population, 3))
print("H s tng quan gia Dn s 2022 v Mt  dn s:", round(corr_population_density, 3))
```
